# Implemeting Root Mean Sqaure Error

```python
def rootMeanSquareError():  
    y_true = np.array([3.0, -0.5, 2, 7])
    y_pred = np.array([2.5, 0.0, 2, 8])
    mse = (np.square(y_true - y_pred)).mean()
    rmse = np.sqrt(mse)
    print(rmse)
    print("Confirmation: ")
    print(root_mean_squared_error(y_true, y_pred))

rootMeanSquareError()
```

# Implementing Z-Indexing

```python
def Zindexing():
    mat = np.random.randint(1,100, (7,5))
    mean = mat.mean(axis = 0)
    std = mat.std(axis = 0)

    mat_Z_indexed = (mat - mean) / std
    print(mat_Z_indexed)

Zindexing()
```

```python
def linearRegression():

    # generate synthetic linear data: y = 4 + 3x + noise
    X = np.random.randn(100, 1)
    y = 4 + 3 * X + np.random.randn(100, 1) * 0.5

    sns.regplot(x = X, y=y)
    plt.show()

    X_biased = np.c_[np.ones((X.shape[0], 1)), X]
    print(X_biased)

    thetaBest = np.linalg.inv(X_biased.T @ X_biased) @ X_biased.T @ y
    print(thetaBest)
```

```python
linearRegression()
```

```python
import numpy as np
import time

# Data
X = np.array([[1, 2], [2, 3], [4, 5], [3, 1], [5, 2]])  # (5,2)
y = np.array([[5], [8], [14], [5], [9]])  # (5,1)

# Parameters
np.random.seed(0)
W = np.random.randn(2, 1)  # (2,1)
b = np.zeros((1,))         # scalar

# Hyperparams
lr = 0.01
epochs = 10000

for i in range(epochs):
    # Forward pass
    y_pred = X @ W + b               # (52)@(21) + (1,) -> (51)
    
    # Loss
    loss = (1/X.shape[0]) * np.sum((y_pred - y) ** 2)
    
    # Gradients
    dW = (2/X.shape[0]) * X.T @ (y_pred - y)  # (25)@(51) = (21)
    db = (2/X.shape[0]) * np.sum(y_pred - y)  # scalar
    
    # Update
    W -= lr * dW
    b -= lr * db
    
    # Print every 100 epochs
    if i % 100 == 0:
        time.sleep(1)
        print(f"Epoch {i}, Loss: {loss:.4f}")
        print(f"{np.array([[7,2]]) @ W + b}")
        print(f"Weight: {W}, bias: {b}")
        print('--------------------')
```
